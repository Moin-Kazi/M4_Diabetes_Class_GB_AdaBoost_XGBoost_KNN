```python
diabet['glucose_conc'] = np.where(diabet['glucose_conc']==0,diabet['glucose_conc'].median(),
                                  diabet['glucose_conc'])
```

```python
diabet['blood_pressure'] = np.where(diabet['blood_pressure']==0,diabet['blood_pressure'].median(),
                                  diabet['blood_pressure'])
```

```python
diabet['skin_fold_thickness'] = np.where(diabet['skin_fold_thickness']==0,diabet['skin_fold_thickness'].median(),
                                  diabet['skin_fold_thickness'])
```

```python
diabet['2-Hour_serum_insulin'] = np.where(diabet['2-Hour_serum_insulin']==0,diabet['2-Hour_serum_insulin'].median(),
                                  diabet['2-Hour_serum_insulin'])
```

```python
diabet['BMI'] = np.where(diabet['BMI']==0,diabet['BMI'].median(),
                                  diabet['BMI'])
```

```python
diabet.describe()
```

```python
diabet.head()
```

```python
diabet['Is_Diabetic'] = diabet['Is_Diabetic'].astype('category')
diabet['Is_Diabetic'] = diabet['Is_Diabetic'].cat.codes
```

```python
diabet.head()
```

```python
diabet['Is_Diabetic'].value_counts()
```

```python
x = diabet.iloc[:,:-1]
y = diabet.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test, y_train,y_test = train_test_split(x,y,train_size=0.7, random_state=101)
```

```python
# Logistic Regression
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
y_pred_lr_train = lr.predict(x_train)
y_pred_lr_test = lr.predict(x_test)
```

```python
print(accuracy_score(y_train, y_pred_lr_train))
print("######"*20)
print(accuracy_score(y_test, y_pred_lr_test))
```

```python
# DecisionTree
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
y_pred_dt_train = dt.predict(x_train)
y_pred_dt_test = dt.predict(x_test)
```

```python
print(accuracy_score(y_train, y_pred_dt_train))
print("######"*10)
print(accuracy_score(y_test, y_pred_dt_test))
```

```python
# Bagging
bg = BaggingClassifier()
bg.fit(x_train,y_train)
```

```python
y_pred_bg_train = bg.predict(x_train)
y_pred_bg_test = bg.predict(x_test)
```

```python
print(accuracy_score(y_train, y_pred_bg_train))
print("######"*10)
print(accuracy_score(y_test, y_pred_bg_test))
```

```python
# RandomForest
rf = RandomForestClassifier(n_estimators=500,criterion='entropy',random_state=551)
rf.fit(x_train, y_train)
```

```python
y_pred_rf_train = rf.predict(x_train)
y_pred_rf_test = rf.predict(x_test)
```

```python
print(accuracy_score(y_train, y_pred_rf_train))
print("######"*10)
print(accuracy_score(y_test, y_pred_rf_test))
```

```python
# cross validation approach to handle overfitting problem in random forest
```

```python
from sklearn.model_selection import cross_val_score
accuracy1 = cross_val_score(rf, x_train, y_train, cv=20)
accuracy2 = cross_val_score(rf, x_test, y_test, cv=20)
print(accuracy1)
print("###"*10)
print(accuracy2)
```
